```python
import pandas as pd 
import numpy as np
```

```python
X_test = pd.read_csv("data/test.csv")
```

```python
train = pd.read_csv("data/train.csv")
test = pd.read_csv("data/test.csv")
```

```python
y_train = train.iloc[:,1].values
X_train = train.drop("Survived",axis=1)
```

```python
X_train = X_train.drop("Name",axis=1)
```

```python
X_train = X_train.drop("PassengerId",axis=1)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder , LabelEncoder

#num_cols = X_train.select_dtypes(include=[np.number]).columns
#cat_cols = X_train.select_dtypes(include=["objects"]).columns

num_cols = ["Age","SibSp","Parch","Fare"]
cat_cols = ["Pclass","Sex","Ticket","Cabin","Embarked"]

num_pipeline = Pipeline(steps=[
    ("imputer",SimpleImputer(strategy="mean"))
])

cat_pipeline = Pipeline(steps=[
    ("imputer",SimpleImputer(strategy="most_frequent")),
    ("onehot",OneHotEncoder(handle_unknown="ignore"))
])

ct = ColumnTransformer(
    transformers=[
        ("num",num_pipeline,num_cols),
        ("cat",cat_pipeline,cat_cols)
    ]
)
```

```python
from sklearn.linear_model import LogisticRegression

model = Pipeline(steps=[
    ("preprocess",ct),
    ("log_reg",LogisticRegression(max_iter=1000))
])
```

```python
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
Subs = pd.read_csv("Submission.csv")
```

```python
Subs.iloc[:,-1] = model.predict(X_test)
```

```python
Subs.to_csv("Submission.csv",index=False)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = Pipeline(steps=[
    ("preprocess",ct),
    ("rfr",RandomForestRegressor())
])
```

```python
model.fit(X_train,y_train)
```

```python
rfr_pred = model.predict(X_test)
```

```python
pred_class = (rfr_pred > 0.5).astype(int)
```

```python
pred_class
```

```python
Subs.iloc[:,-1] = pred_class
Subs.to_csv("Submission.csv",index=False)
```
